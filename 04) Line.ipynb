```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])
```

```python
plt.plot(ypoints, linestyle = 'dotted',color='r')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed',color='hotpink',linewidth='10')
plt.show()
```

```python
plt.plot(ypoints, ls = ':')
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
